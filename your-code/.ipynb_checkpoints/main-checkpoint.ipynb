####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html=requests.get(url)
#print(html)
html=requests.get(url).content
#print(html)
soup=BeautifulSoup(html,'lxml')

#INTENTO 1
data=soup.select('b > a')
data
data_clean=[element.text.strip() for element in data]
data_clean
#cuando imprimía data_clean, se imprime la lista con \xa0 pero en el df, ya no

latitude=soup.select('tr td.tabev1') #'tr >td:nth-child(6)') le quise agregar si era N o S  y me salió \nTypeError: ABCMeta object argument after ** must be a mapping, not str
latitude
latitude_clean=[element.text.strip() for element in latitude]
latitude_clean

longitude=soup.select('tr >td:nth-child(7)')#,'tr >td:nth-child(8)') le quise agregar si era W o E  y me salió \nTypeError: ABCMeta object argument after ** must be a mapping, not str
longitude
longitude_clean=[element.text.strip() for element in longitude]
longitude_clean

region=soup.select('tr td.tb_region')
region
region_clean=[element.text.strip() for element in region]
region_clean

df=pd.DataFrame(list(zip(data_clean,latitude_clean,longitude_clean,region_clean)),columns=['Data_and_time','Latitude','Longitude','Region'])
df.head(20)
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
url ='https://mlh.io/seasons/na-2020/events'
````

````python
#your code
html=requests.get(url)
#print(html)
html=requests.get(url).content
#print(html)
soup=BeautifulSoup(html,'lxml')

date=soup.select('div p.event-date')
date
date_clean=[element.text.strip() for element in date]
date_clean

city=soup.select('div>span:nth-child(1)')
city
city_clean=[element.text.strip() for element in city]
city_clean

country=soup.select('div>span:nth-child(2)')
country
country_clean=[element.text.strip() for element in country]
country_clean

title=soup.select('div h3.event-name')
title
title_clean=[element.text.strip() for element in title]
title_clean

df=pd.DataFrame(list(zip(date_clean,city_clean,country_clean,title_clean)),columns=['Date','City','Country','Title'])
df.head(25)
````
